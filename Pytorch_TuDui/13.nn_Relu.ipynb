### 非线性激活函数,RELU,Sigmoid

```python
import torch
import torchvision
from torch import nn
from torch.nn import Conv2d
from torch.utils.data import DataLoader
from torch.utils.tensorboard import SummaryWriter
```

```python
from torch.nn import ReLU
from torch.nn import Sigmoid
```

```python
dataset = torchvision.datasets.CIFAR10(root="./dataset",train=False,transform=torchvision.transforms.ToTensor())
```

```python
dataloader = DataLoader(dataset,batch_size=64)
```

```python
writer = SummaryWriter("logs")
```

模型

```python
class TuDui(nn.Module):
    def __init__(self):
        super(TuDui,self).__init__()
        self.relu = ReLU()
        self.sigmoid = Sigmoid()

    def forward(self,input,activate):
        if(activate == 'ReLU'):
            output = self.relu(input)
        if(activate == 'Sigmoid'):
            output = self.sigmoid(input)
        return output
```

```python
tudui = TuDui()
```

```python
input = torch.tensor([[1, -0.5],
                      [-1, 3]])

input = torch.reshape(input, (-1, 1, 2, 2))
```

```python
output = tudui.forward(input,'Sigmoid')
```

由于图像中没有负值，所以Relu函数，不改变图像

```python
step = 0
for data in dataloader:
    imgs,targets = data
    writer.add_images("source",imgs,step)
    output_relu = tudui.forward(imgs,'ReLU')
    writer.add_images("ReLU",output_relu,step)
    output_sigmoid = tudui.forward(imgs,'Sigmoid')
    writer.add_images("Sigmoid",output_sigmoid,step)
    step += 1
```

```python
writer.close()
```
